### **Normal Distribution Analysis**

---



You work as a business analyst in a sports company that sells sports shoes. For our customer group, we know that the fitting shoe length is normally distributed with a mean (μ) of 27 cm and a standard deviation (σ) of 2.5 cm.

We want to choose between 2 variants for the model that fits more people:
1. **Model A**: Fits for shoe length range 27.1 cm - 27.6 cm
2. **Model B**: Fits for shoe length range 26.7 cm - 27.1 cm

To determine the percentage of customers that fit each shoe model, we'll use the properties of the normal distribution with a mean (μ) of 27 cm and a standard deviation (σ) of 2.5 cm.

### **Model A: Fits for shoe length range 27.1 cm - 27.6 cm**

1. Calculate the z-scores for 27.1 cm and 27.6 cm:

z1 = {27.1 - 27}/{2.5} = {0.1}/{2.5} = 0.04

z2 = {27.6 - 27}/{2.5} = {0.6}/{2.5} = 0.24

2. Find the CDF values for these z-scores using standard normal distribution tables or a calculator:


P(Z <= 0.04) \approx 0.5160

P(Z <= 0.24) \approx 0.5948

3. Calculate the percentage of customers that fit Model A:

P(27.1 <= X <= 27.6) = P(Z <= 0.24) - P(Z <= 0.04) = 0.5948 - 0.5160 = 0.0788

So, approximately \(7.88\%\) of customers fit Model A.

### **Model B: Fits for shoe length range 26.7 cm - 27.1 cm**

1. Calculate the z-scores for 26.7 cm and 27.1 cm:

z3 = {26.7 - 27}/{2.5} = {-0.3}/{2.5} = -0.12

z1 = 0.04 {(already calculated for Model A)}

2. Find the CDF values for these z-scores using standard normal distribution tables or a calculator:

P(Z <= -0.12) \approx 0.4522

P(Z <= 0.04) \approx 0.5160 {(already calculated for Model A)}


3. Calculate the percentage of customers that fit Model B:

P(26.7 <= X <= 27.1) = P(Z <= 0.04) - P(Z <= -0.12) = 0.5160 - 0.4522 = 0.0638


So, approximately \(6.38\%\) of customers fit Model B.

### **Summary:**

- **Model A:** Fits approximately \(7.88\%\) of customers.
- **Model B:** Fits approximately \(6.38\%\) of customers.

---



```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Mean and standard deviation
mu = 27
sigma = 2.5

# Calculate z-scores for Model A
z1_A = (27.1 - mu) / sigma
z2_A = (27.6 - mu) / sigma

# Calculate CDF values for Model A
p1_A = stats.norm.cdf(z1_A)
p2_A = stats.norm.cdf(z2_A)

# Calculate the percentage of customers that fit Model A
percent_A = (p2_A - p1_A) * 100

# Calculate z-scores for Model B
z1_B = (26.7 - mu) / sigma
z2_B = (27.1 - mu) / sigma

# Calculate CDF values for Model B
p1_B = stats.norm.cdf(z1_B)
p2_B = stats.norm.cdf(z2_B)

# Calculate the percentage of customers that fit Model B
percent_B = (p2_B - p1_B) * 100

# Print the results
print(f"Percentage of customers that fit Model A: {percent_A:.2f}%")
print(f"Percentage of customers that fit Model B: {percent_B:.2f}%")

# Plotting the normal distribution and the ranges for Model A and Model B
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = stats.norm.pdf(x, mu, sigma)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Normal Distribution')
plt.fill_between(x, 0, y, where=(x >= 27.1) & (x <= 27.6), color='blue', alpha=0.3, label='Model A Range')
plt.fill_between(x, 0, y, where=(x >= 26.7) & (x <= 27.1), color='green', alpha=0.3, label='Model B Range')
plt.title('Normal Distribution of Shoe Lengths with Model A and Model B Ranges')
plt.xlabel('Shoe Length (cm)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
